accuracy/confusion-matrix/PRF requirements. Each item the brief asks for is present,

## Dataset class

```python

def encode_labels(label_str, target=TARGET):
    """Multi-hot encode a pipe-separated label string into a float32 vector."""
    labels = set(l.strip() for l in label_str.split('|'))
    return [1.0 if t in labels else 0.0 for t in target]


# Build disk index ONCE, shared across all three datasets
print('Indexing images on disk (one-time scan)...')
_t = time.time()
IMG_INDEX = {p.name: str(p) for p in Path(IMG_DIR).rglob('*.png')}
print(f'  {len(IMG_INDEX):,} images indexed in {time.time()-_t:.1f}s')


class ChestXrayDataset(Dataset):
    def __init__(self, csv_path, img_dir, transform=None, img_index=IMG_INDEX, cache=True):
        self.df        = pd.read_csv(csv_path)
        self.transform = transform

        label_matrix = self.df['Finding Labels'].apply(encode_labels).tolist()
        self.labels  = torch.tensor(label_matrix, dtype=torch.float32)

        self.df['img_path'] = self.df['Image Index'].map(img_index)
        valid        = self.df['img_path'].notna()
        self.df      = self.df[valid].reset_index(drop=True)
        self.labels  = self.labels[valid.values]
        self.paths   = self.df['img_path'].tolist()
        print(f'  {Path(csv_path).name}: {len(self.df):,} images matched')

        # Pre-cache images in RAM 
        # self.cache = {}
        # if cache:
        #     print(f'  Caching {len(self.paths):,} images in RAM...')
        #     for i, p in enumerate(tqdm(self.paths, leave=False)):
        #         self.cache[i] = Image.open(p).convert('RGB').resize(
        #             (CACHE_SIZE, CACHE_SIZE), Image.BILINEAR
        #         )
        #     print(f'Cache ready')

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        # img = self.cache[idx] if self.cache else Image.open(self.paths[idx]).convert('RGB')
        img = Image.open(self.paths[idx]).convert('RGB')
        if self.transform:
            img = self.transform(img)
        return img, self.labels[idx]


print('\nLoading datasets...')
train_ds = ChestXrayDataset(PROC_CX / 'train.csv', IMG_DIR, train_transform)
val_ds   = ChestXrayDataset(PROC_CX / 'val.csv',   IMG_DIR, val_transform)
# test_ds  = ChestXrayDataset(PROC_CX / 'test.csv',  IMG_DIR, val_transform)
# print(f'\nTrain: {len(train_ds):,}  |  Val: {len(val_ds):,}  |  Test: {len(test_ds):,}')
print(f'\n  Test: {len(train_ds):,}')
```

## Patient-level split check

ChestX-ray14 has multiple images per patient; if a patient spans splits, test AUC is
inflated by leakage. The preprocessing notebook split by `Patient ID`; we re-verify here
(a clean ✅ is good evidence to show on the poster).

```python
def check_patient_leakage(*csv_paths):
    sets, names = [], []
    for p in csv_paths:
        df = pd.read_csv(p)
        col = next((c for c in df.columns if c.strip().lower()=='patient id'), None)
        if col is None:
            print(f'  ⚠️  {Path(p).name}: no "Patient ID" column — verify split is patient-wise.')
            return
        sets.append(set(df[col].unique())); names.append(Path(p).name)
    leaks = False
    for a in range(len(sets)):
        for b in range(a+1, len(sets)):
            ov = sets[a] & sets[b]
            if ov:
                leaks = True
                print(f'  ❌ {len(ov)} patients shared between {names[a]} & {names[b]}')
    if not leaks:
        print('  ✅ No patient overlap across splits — clean split.')

check_patient_leakage(PROC_CX/'train.csv', PROC_CX/'val.csv', PROC_CX/'test.csv')
```

# WeightedRandomSampler + DataLoaders
 Per-sample weight = inverse frequency of its rarest positive label.   
This handles the severe 14-class imbalance (No Finding 60k vs Hernia 227)   
without relying on TARGET column lookups — uses the pre-built label tensors.   

```python
def make_sampler(dataset):
    label_matrix  = dataset.labels.numpy()                  # (N, 15)
    class_counts  = label_matrix.sum(axis=0)                # (15,) positives per class
    class_weights = 1.0 / np.maximum(class_counts, 1)       # inverse freq, avoid div/0

    sample_weights = []
    for row in label_matrix:
        pos_weights = row * class_weights                    # zero out negatives
        # weight = rarest positive class; fall back to min weight if all-negative
        w = pos_weights.max() if pos_weights.max() > 0 else class_weights.min()
        sample_weights.append(w)

    return WeightedRandomSampler(
        torch.DoubleTensor(sample_weights),
        num_samples=len(sample_weights),
        replacement=True
    )


# Single imbalance correction only: pos_weight in the loss (below). We do NOT also
# use the WeightedRandomSampler — stacking both double-counts the correction, over-
# predicts rare findings and caps AUC. make_sampler is left defined for reference.
# train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
#                           num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=True)

# print(f'Train batches : {len(train_loader)}')
print(f'Val batches   : {len(val_loader)}')
# print(f'Test batches  : {len(test_loader)}')
print('DataLoaders ready')
```

# Asymmetric Loss

## Loss Function
BCEWithLogitsLoss with pos_weight balances per-class imbalance.   
pos_weight[i] = neg_count[i] / pos_count[i]: the loss for a positive   
example of class i is scaled up proportionally to how rare it is,   
so Hernia (227 samples) is penalised ~263× more than No Finding (60k).   

```python
def compute_pos_weight(dataset, device):
    """Compute per-class pos_weight = neg_count / pos_count for BCEWithLogitsLoss."""
    label_matrix = dataset.labels.numpy()          # (N, 15)
    pos_counts   = label_matrix.sum(axis=0)        # positives per class
    neg_counts   = len(label_matrix) - pos_counts  # negatives per class
    pos_weight   = neg_counts / np.maximum(pos_counts, 1)  # avoid div/0
    pos_weight   = np.minimum(pos_weight, 20.0)            # cap (Hernia ~263x) to keep BCE stable
    return torch.tensor(pos_weight, dtype=torch.float32).to(device)


pos_weight = compute_pos_weight(train_ds, DEVICE)
criterion  = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

print('BCEWithLogitsLoss with pos_weight ready')
print(f'  pos_weight range: {pos_weight.min():.1f} (No Finding) → {pos_weight.max():.1f} (capped 20)')
print(f'  Classes: {list(zip(TARGET, pos_weight.cpu().numpy().round(1)))}')
```

# Models — three pretrained backbones from `timm`

All three models are created with `pretrained=True` and `num_classes=15`, which gives
each an identically-shaped classification head sized to our label space. We apply the
sigmoid at the **loss** (`BCEWithLogitsLoss`), so the head is the same across all three —
a fair comparison. `timm` handles the ImageNet weights and architecture details.

A factory returns a fresh model on demand so each is trained independently.

```python
def build_model(timm_name, num_classes=len(TARGET)):
    """Create a pretrained timm model with a fresh multi-label head."""
    model = timm.create_model(timm_name, pretrained=True, num_classes=num_classes)
    return model

# Map friendly names -> timm identifiers (matches the competition's required models)
MODEL_ZOO = {
    'ResNet50V2':       'resnetv2_50',
    'ViT_B16':          'vit_base_patch16_224',
    'EfficientNetV2B0': 'tf_efficientnetv2_b0',
}

# Quick parameter count sanity check (no weights downloaded twice — cached after first use)
for name, tn in MODEL_ZOO.items():
    m = build_model(tn)
    n = sum(p.numel() for p in m.parameters())
    print(f'  {name:18s} ({tn:22s}) : {n/1e6:6.1f}M params')
    del m
print('Model factory ready')
```

## Backbone freeze / unfreeze helpers

Two-phase fine-tuning needs to freeze the pretrained backbone (train only the head) and
later unfreeze it. `timm` models expose the classifier via `get_classifier()`, so we
freeze everything then re-enable just the head for phase 1.

```python
def freeze_backbone(model):
    """Freeze all params except the classifier head (phase 1)."""
    for p in model.parameters():
        p.requires_grad = False
    head = model.get_classifier()
    for p in head.parameters():
        p.requires_grad = True

def unfreeze_all(model):
    """Unfreeze every parameter (phase 2 fine-tuning)."""
    for p in model.parameters():
        p.requires_grad = True

print('Freeze/unfreeze helpers ready')
```

# Training pipeline — two-phase transfer learning

Each model is trained in two phases: **(1)** backbone frozen, train the head at `HEAD_LR`;
**(2)** unfreeze everything and fine-tune at the low `FT_LR`. We use mixed precision,
checkpoint on best **validation AUC**, and keep full history for the required learning
curves. A `_full.pth` checkpoint is written every epoch for exact resume.

```python
# Training & evaluation functions

def train_one_epoch(model, loader, criterion, optimizer, scaler, device):
    model.train()
    total_loss = 0
    for imgs, labels in tqdm(loader, desc='  Train', leave=False):
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        with torch.amp.autocast('cuda'):
            loss = criterion(model(imgs), labels)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        total_loss += loss.item()
    return total_loss / len(loader)


@torch.no_grad()
def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, all_probs, all_labels = 0, [], []
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        logits       = model(imgs)
        total_loss  += criterion(logits, labels).item()
        all_probs.append(torch.sigmoid(logits).cpu().numpy())
        all_labels.append(labels.cpu().numpy())
    probs  = np.vstack(all_probs)
    labels = np.vstack(all_labels)
    aucs   = []
    for i in range(labels.shape[1]):
        if labels[:, i].sum() > 0:
            aucs.append(roc_auc_score(labels[:, i], probs[:, i]))
        else:
            aucs.append(float('nan'))
    return total_loss / len(loader), np.nanmean(aucs), aucs, probs, labels


def _run_phase(model, name, phase_tag, loader, val_loader, criterion,
               optimizer, scheduler, scaler, epochs, device, ckpt_dir,
               best_auc, history):
    """Run one training phase; update best checkpoint on val AUC."""
    best_weights = None
    patience_cnt, patience = 0, 6
    for epoch in range(1, epochs + 1):
        t0 = time.time()
        tr = train_one_epoch(model, loader, criterion, optimizer, scaler, device)
        vl, va, _, _, _ = evaluate(model, val_loader, criterion, device)
        cur_lr = optimizer.param_groups[0]['lr']
        if scheduler is not None:
            scheduler.step()
        history['train_loss'].append(tr); history['val_loss'].append(vl)
        history['val_auc'].append(va);    history['lr'].append(cur_lr)
        print(f'  [{phase_tag}] Epoch {epoch:02d}/{epochs} | Train {tr:.4f} | '
              f'Val {vl:.4f} | AUC {va:.4f} | LR {cur_lr:.2e} | {time.time()-t0:.0f}s')
        if va > best_auc:
            best_auc = va
            best_weights = copy.deepcopy(model.state_dict())
            torch.save(best_weights, ckpt_dir / f'{name}_best.pth')
            print(f'         ✅ New best AUC: {best_auc:.4f} — saved')
            patience_cnt = 0
        else:
            patience_cnt += 1
        torch.save({'model': model.state_dict(), 'optimizer': optimizer.state_dict(),
                    'scaler': scaler.state_dict(), 'best_auc': best_auc,
                    'history': history}, ckpt_dir / f'{name}_full.pth')
        if patience_cnt >= patience:
            print(f'         ⏹️  Early stopping ({phase_tag}) at epoch {epoch}')
            break
    return best_auc, best_weights


def train_model(model, name, train_loader, val_loader, criterion, device, ckpt_dir,
                head_epochs=HEAD_EPOCHS, ft_epochs=FT_EPOCHS,
                head_lr=HEAD_LR, ft_lr=FT_LR):
    """Two-phase transfer learning: frozen-head warm-up, then full fine-tune."""
    model = model.to(device)
    scaler  = torch.amp.GradScaler('cuda')
    history = {'train_loss': [], 'val_loss': [], 'val_auc': [], 'lr': []}
    best_auc, best_weights = 0.0, None

    print(f'\n{"="*60}\n  Training {name}\n{"="*60}')

    # ---- Phase 1: frozen backbone, train head ----
    freeze_backbone(model)
    opt = optim.AdamW([p for p in model.parameters() if p.requires_grad],
                      lr=head_lr, weight_decay=1e-4)
    print(f'  Phase 1 — head only ({head_epochs} epochs @ {head_lr:g})')
    best_auc, bw = _run_phase(model, name, 'head', train_loader, val_loader,
                              criterion, opt, None, scaler, head_epochs,
                              device, ckpt_dir, best_auc, history)
    if bw is not None: best_weights = bw

    # ---- Phase 2: unfreeze + fine-tune with cosine decay ----
    unfreeze_all(model)
    opt = optim.AdamW(model.parameters(), lr=ft_lr, weight_decay=1e-4)
    sched = optim.lr_scheduler.CosineAnnealingLR(opt, T_max=ft_epochs)
    print(f'  Phase 2 — fine-tune all ({ft_epochs} epochs @ {ft_lr:g}, cosine)')
    best_auc, bw = _run_phase(model, name, 'ft', train_loader, val_loader,
                              criterion, opt, sched, scaler, ft_epochs,
                              device, ckpt_dir, best_auc, history)
    if bw is not None: best_weights = bw

    if best_weights is not None:
        model.load_state_dict(best_weights)
    print(f'\n  ✅ {name} done | Best Val AUC: {best_auc:.4f}')
    return model, history


print('Training functions defined')
```

# Train all three models

Each model is built fresh from the timm factory and trained with the two-phase schedule.
Histories and best checkpoints are stored per model for the comparison section.
On 8 GB VRAM, if ViT fine-tuning OOMs: set `BATCH_SIZE = 16` in the config cell and re-run.

```python
models, histories = {}, {}
for name, timm_name in MODEL_ZOO.items():
    net = build_model(timm_name)
    net, hist = train_model(net, name, train_loader, val_loader,
                            criterion, DEVICE, CKPT_DIR)
    models[name], histories[name] = net, hist

print('\nAll models trained:', list(models))
```

### Resume note
To resume an interrupted model, rebuild it with `build_model(MODEL_ZOO[name])`, load
`CKPT_DIR/f'{name}_full.pth'`, and restore the model/optimizer/scaler before continuing.
For a clean re-run, just re-execute the training loop above (best checkpoints overwrite).

```python
def load_histories(ckpt_dir):
    """Rebuild the {name: history} dict from saved *_full.pth checkpoints."""
    ckpt_dir = Path(ckpt_dir)
    histories = {}
    for name in MODEL_ZOO:                       # keeps MODEL_ZOO ordering
        full_path = ckpt_dir / f'{name}_full.pth'
        if not full_path.exists():
            print(f'  ⏭️  {name:18s} — no checkpoint, skipping')
            continue
        ckpt = torch.load(full_path, map_location='cpu', weights_only=False)
        hist = ckpt.get('history')
        if hist is None:
            print(f'  ⚠️  {name:18s} — checkpoint has no history')
            continue
        histories[name] = hist
        best_auc = ckpt.get('best_auc', float('nan'))
        print(f'  ✅ {name:18s} | epochs logged: {len(hist["val_auc"]):2d} | '
              f'best AUC {best_auc:.4f}')
    return histories


def load_models(ckpt_dir, device):
    """Rebuild each model from MODEL_ZOO and load its best weights."""
    ckpt_dir = Path(ckpt_dir)
    models = {}
    for name, timm_name in MODEL_ZOO.items():
        best_path = ckpt_dir / f'{name}_best.pth'
        if not best_path.exists():
            print(f'  ⏭️  {name:18s} — no checkpoint, skipping')
            continue
        model = timm.create_model(timm_name, pretrained=False,
                                  num_classes=len(TARGET)).to(device)
        state = torch.load(best_path, map_location=device)   # weights_only=True ok
        model.load_state_dict(state)
        model.eval()
        models[name] = model
        print(f'  ✅ {name:18s} loaded')
    return models


models = load_models(CKPT_DIR, DEVICE)
print('\nLoaded models:', list(models))


histories = load_histories(CKPT_DIR)
print('\nLoaded histories for:', list(histories))
```

```python
# Test-set evaluation + validation-tuned thresholds
def tune_thresholds(y_true, y_prob):
    ths = np.full(y_true.shape[1], 0.5)
    for i in range(y_true.shape[1]):
        if y_true[:, i].sum() == 0: continue
        best_f1, best_t = 0.0, 0.5
        for t in np.linspace(0.05, 0.95, 19):
            _,_,f1,_ = precision_recall_fscore_support(
                y_true[:, i], (y_prob[:, i] >= t).astype(int),
                average='binary', zero_division=0)
            if f1 > best_f1: best_f1, best_t = f1, t
        ths[i] = best_t
    return ths

results = {}
print('Evaluating on TEST set ...')
for name, model in models.items():
    _, _, _, val_prob, val_lab   = evaluate(model, val_loader,  criterion, DEVICE)
    _, mean_auc, aucs, te_prob, te_lab = evaluate(model, test_loader, criterion, DEVICE)
    ths  = tune_thresholds(val_lab, val_prob)
    pred = (te_prob >= ths).astype(int)
    p, r, f1, _   = precision_recall_fscore_support(te_lab, pred, average=None,   zero_division=0)
    _, _, macro_f1, _ = precision_recall_fscore_support(te_lab, pred, average='macro', zero_division=0)
    results[name] = dict(mean_auc=mean_auc, aucs=aucs, precision=p, recall=r, f1=f1,
                         macro_f1=macro_f1, thresholds=ths, probs=te_prob,
                         labels=te_lab, preds=pred)
    print(f'  {name:18s} | mean AUC {mean_auc:.4f} | macro-F1 {macro_f1:.4f}')
```

# Responsible deployment, system design & ethics (required)

The brief's Section 2 was written for *facial video*; re-grounded for chest X-rays:

**System architecture (frontend / middleware / backend).**
- **Frontend** — clinician-facing viewer (web app or PACS workstation integration) that
  pulls a study and shows model outputs as *decision support*, never autonomous diagnosis.
  Tools: React/Vue, OHIF viewer, DICOMweb.
- **Middleware** — orchestration/inference layer: receives the DICOM study, de-identifies,
  preprocesses (windowing, resize, grayscale→RGB), batches to the model server, applies
  thresholds, returns structured findings. Tools: FastAPI, a queue (RabbitMQ/Kafka),
  `pydicom`, an inference server (TorchServe / Triton).
- **Backend** — model serving + results/audit store with monitoring & governance logging.
  Tools: TorchServe/Triton, Postgres, object storage, MLflow for model versioning.

**"Temporal smoothing" — the clinical analogue.** Single X-rays have no video frames, so
frame-to-frame smoothing does not apply directly. Equivalent stabilisers: (a) aggregating
a patient's **multiple studies over time** (prior comparison), (b) **test-time augmentation**
(averaging predictions over flips/crops), and (c) **ensembling** the three models to damp
single-model variance — all producing more stable, reliable outputs.

**Privacy, personal & biometric data (GDPR).** Chest X-rays are **health data** —
*special-category* personal data under **GDPR Article 9**, with stricter conditions than
ordinary personal data; medical images can also be **biometric data** where used to
identify individuals. Implications: a lawful basis plus an Article 9 condition (e.g.
explicit consent or provision of healthcare), a **DPIA**, robust DICOM de-identification
(strip PHI tags), data minimisation, encryption in transit and at rest, access control,
audit logging, and clear retention/deletion policies. Models should be checked for
**bias across demographic subgroups** before deployment, with outputs framed as assistive
to an accountable clinician.

**Practical deployment challenges.** Distribution shift across scanners/sites; the
label-noise ceiling on achievable performance; per-population threshold calibration;
medical-device software regulation; and avoiding alert fatigue in the clinical workflow.
